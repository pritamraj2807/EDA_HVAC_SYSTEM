SPLITING THE TEST AND TRAIN DATA

```
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data2, test_size=0.2, random_state=42)
```

```
X_train = train_set.drop('HVAC System: Electricity', axis = 1)
X_test = test_set.drop('HVAC System: Electricity', axis = 1)
y_train = train_set['HVAC System: Electricity']
y_test = test_set['HVAC System: Electricity']
```

```
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```
X_train_numerical = X_train.loc[:,X_train.dtypes==np.float64]
X_train_cat = X_train.loc[:,X_train.dtypes==np.int64]
X_test_numerical = X_test.loc[:,X_test.dtypes==np.float64]
X_test_cat = X_test.loc[:,X_test.dtypes==np.int64]
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_numerical = scaler.fit_transform(X_train_numerical)
X_test_numerical = scaler.fit_transform(X_test_numerical)
```

```
X_train_cat = np.array(X_train_cat)
X_test_cat = np.array(X_test_cat)
```

```
X_train = np.hstack([X_train_numerical, X_train_cat])
X_test = np.hstack([X_test_numerical, X_test_cat])
```

RANDOM FOREST

```
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import RandomForestRegressor
```

```
rf_regressor = RandomForestRegressor()
y_train_predict = cross_val_predict(rf_regressor, X_train, y_train, cv = 3, n_jobs = -1)
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```
mse = mean_squared_error(y_train, y_train_predict)
```

```
rmse = np.sqrt(mse)
```

```
rmse
```

```
mae = mean_absolute_error(y_train, y_train_predict)
mae
```
